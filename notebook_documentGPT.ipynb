# 4.1 FewShotPromptTemplate
## 템플릿을 만들면 좋은 점
- template 값을 넘겼는지 validation
- save and load prompt template to the disk.

## FewShotPromptTemplate
- 모델이 어떻게 대답해야 하는지 예제를 준다.
- 기존의 데이터베이스에 있는 예제들을 학습시킬 때 좋다. 예를 들어 고객지원.
- 사용법
    1. 예제 리스트를 준비한다.
    2. 예제를 format할 prompt를 만든다.
    3. fewShotPromptTemplate에 다음과 같은 parameters를 전달한다.
        - example_prompt = 예제 형식
        - examples = 예제
        - suffix = 실제 질문
        - input_variables = ["유효성 검사를 받을 수 있도록, 템플릿에서 사용할 입력 변수가 무엇인지 지정"]
    4. 체인에 넣는다.

## FewShotPromptTemplate과 ChatPromptTemplate 용도의 차이점
- FewShotPromptTemplate: Format example conversations by insert multiple example chat pairs.
- ChatPromptTemplate: Build the final full prompt to assemble system + examples + real user input

1. Random Example Selector

````python
'''
from langchain.chat_models import ChatOpenAI
from langchain.prompts import PromptTemplate, FewShotPromptTemplate, example_selector
from langchain.prompts.example_selector.base import BaseExampleSelector
from langchain.callbacks import StreamingStdOutCallbackHandler


chat = ChatOpenAI(
    temperature=0.1,
    streaming=True,
    callbacks=[StreamingStdOutCallbackHandler()],
)

examples = [
    {
        "question": "What do you know about France?",
        "answer": """
        Here is what I know:
        Capital: Paris
        Language: French
        Food: Wine and Cheese
        Currency: Euro
        """,
    },
    {
        "question": "What do you know about Italy?",
        "answer": """
        I know this:
        Capital: Rome
        Language: Italian
        Food: Pizza and Pasta
        Currency: Euro
        """,
    },
    {
        "question": "What do you know about Greece?",
        "answer": """
        I know this:
        Capital: Athens
        Language: Greek
        Food: Souvlaki and Feta Cheese
        Currency: Euro
        """,
    },
]

class RandomExampleSelector(BaseExampleSelector):
    def __init__(self, examples):
        self.examples = examples

    def add_example(self, example):
        return self.examples.append(example)

    def select_examples(self, input_variables):
        from random import choice
        return [choice(self.examples)]
    
example_prompt = PromptTemplate.from_template("Human: {question}\nAI: {answer}")
example_selector = RandomExampleSelector(examples=examples)

prompt = FewShotPromptTemplate(
    example_prompt=example_prompt,
    example_selector=example_selector,
    suffix="Human: What do you know about {country}?",
    input_variables=["country"],
)

prompt.format(country="Brazil")
'''
````

# 5.0 ConversationBufferMemory ~ 5.4 ConversationKGMemory
## 목표:
1. 각 메모리의 종류와 차이점을 알아본다.
2. 랭체인에 메모리를 탑재하는 법을 배운다.

## 1. ConversationBufferMemory
- Saves whole conversations.
- Cons: 대화가 길어질수록 메모리가 커져 비효율적이다.
- 사용예제: 텍스트 자동완성 시 유용하다.

## 2. ConversationBufferWindowMemory
- 가장 최근 대화만 저장
- Pros: 저장 범위를 지정해 원하는 크기로 메모리를 관리할 수 있다.
- Cons: 챗봇이 전체 대화가 아닌 최근 대화에만 집중한다.

## 3. ConversationSummaryMemory
- Use LLM to make summary of conversation and then save to memory.
- Pros: 대화가 많아질수록 일반 메모리에 비해 토큰을 절감할 수 있다.

## 4. ==ConversationSummaryBufferMemory==
- ConversationSummaryMemory + ConversationBufferMemory
- 메모리에 보낸 메세지수를 세어 최대치를 도달하면 오래된 메세지를 요약한다.

## 5. Conversation Knowledge Graph Memory
- LLM을 사용하여 대화 중 entity의 knowledge graph를 만든다.
- 즉 대화 내용 중 중요한 것만 뽑아 요약본을 만든다.

## 6. Conversation Token Buffer
- ConversationBufferWindowMemory와 비슷한데,
- interaction 최대값 대신 token 총량을 계산한다.

## 7. Entitiy
- 대화 중에 entitiy를 추출

## Custom Chain + Memory Output as Messages.
### Workflow:
1. input: chain.invoke 안에 있는 dictionary가 제일 먼저 실행 -> output: question 구문 출력

3. Put loaded memory to the prompt using chat_history key.
4. Prompt needs the key. 

````python
'''
from langchain.chat_models import ChatOpenAI
from langchain.memory import ConversationSummaryBufferMemory
from langchain.schema.runnable import RunnablePassthrough
from langchain.prompts import ChatPromptTemplate, MessagesPlaceholder

llm = ChatOpenAI(
    temperature=0.1
)

memory = ConversationSummaryBufferMemory(
    llm=llm,
    max_token_limit=120,
    return_messages=True,
)

prompt = ChatPromptTemplate.from_messages([
    ("system","You are a helpful AI talking to human."),
    MessagesPlaceholder(variable_name="history"), # memory가 줄 system, human, ai messages 양이 얼마나 될 지 모를 때 사용
    ("human","{question}"),
])

def load_memory(input): # input={"question": "The Constantinople plays ancient instruments like Kora and Setar."}, _로 무시해도 됨.
    return memory.load_memory_variables({})["history"]

# RunnablePassthrough: chain에 history를 매번 전달하지 않기 위해, prompt가 format 되기 전 함수를 실행
chain = RunnablePassthrough.assign(history=load_memory) | prompt | llm

# 결과를 메모리에 저장하기 위해 체인을 호출하는 함수를 만든다.
def invoke_chain(question):
    result = chain.invoke({"question": question})
    memory.save_context({"input": question}, {"output":result.content}) # result = AI messages' content
    print(result)

invoke_chain("The Constantinople plays ancient instruments like Kora and Setar.")
invoke_chain("What instruments does Constantinople plays?")
'''
````

# 6.8 [Stuff LCEL Chain](https://python.langchain.com/docs/concepts/runnables/)
## 체인의 순서
1. chain.invoke의 query가 chain의 retriever로 전달된다.
2. retriever = input: query와 관련 있는 document를 모두 가져온다. output: List of Documents
3. templates = input: List of Documents가 prompt의 {context}로, query가 human {question}으로 입력된다.

## 아래의 예제에 비추어보면:
chain = (
    {
        "context": retriever("죽은 자가 산 자를 구할 수 있는가?"), -> [Doc]
        "question": RunnablePassthrough("죽은 자가 산 자를 구할 수 있는가?")
    }
    | prompt.format_messages(context=[Doc], question="query") -> formatted prompt
    | llm
)

# 6.9 Map Reduce LCEL Chain
## Map Reduce Process
1. retriever("query from 6.")
2. retriever가 list of documents을 반환한다.
3. for doc in doc_list | doc 마다 각각의 prompt | llm이 doc 수만큼 응답
4. for response in every llm responses_list | put em altogether in one document.
5. {context: the final doc} | prompt | llm("answer the question with these relevant infos.")
6. chain.invoke("query")

## Stuff VS Map Reduce
- retriever가 천 개 이상의 doc을 반환한다면, stuff는 사용할 수 없어. doc이 많은 상황에서 각각의 doc마다 요약하는 Map Reduce 방식을 사용한다.

````python
'''
from langchain.chat_models import ChatOpenAI
from langchain.document_loaders import TextLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.prompts import ChatPromptTemplate
from langchain.schema.runnable import RunnablePassthrough, RunnableLambda
from langchain.storage import LocalFileStore
from langchain.embeddings import OpenAIEmbeddings, CacheBackedEmbeddings
from langchain.vectorstores import FAISS

llm = ChatOpenAI(model="gpt-4o-mini-2024-07-18")
loader = TextLoader("files/han-lecture-korean.txt")
splitter = CharacterTextSplitter.from_tiktoken_encoder(
    separator="*",
    chunk_size=600,
    chunk_overlap=100,
)
docs = loader.load_and_split(text_splitter=splitter)
embeddings = OpenAIEmbeddings()
cache_dir = LocalFileStore(".cache/")
cached_embeddings = CacheBackedEmbeddings.from_bytes_store(embeddings, cache_dir)
vectorstore = FAISS.from_documents(docs, cached_embeddings)
retriever = vectorstore.as_retriever()

# 3. for doc in doc_list | doc 마다 각각의 prompt | llm이 doc 수만큼 응답
map_doc_prompt = ChatPromptTemplate.from_messages([
    (
        "system",
        """
        Use the following portion of a long document to see if any of the text is relevant to answer the question. Return any relevant text verbatim. If there is no relevant text, return : ''
        -------
        {context}
        """,
    ),
])

map_doc_chain = map_doc_prompt | llm

# 2. retriever가 list of documents을 반환한다.
def map_docs(inputs):
    documents = inputs['documents']
    question = inputs['question']
    return "\n\n".join(map_doc_chain.invoke({
        "context": doc.page_content,
        "question": question
    }).content for doc in documents)

map_chain = {
    "documents": retriever, 
    "question": RunnablePassthrough(),
    } | RunnableLambda(map_docs) # document가 질문과 관련이 있는지 확인

# 5. {context: the final doc} | prompt | llm("answer the question with these relevant infos.")
final_prompt = ChatPromptTemplate.from_messages([
    (
        "system",
        """
        Given the following extracted parts of a long document and a question, create a final answer. 
        If you don't know the answer, just say that you don't know. Don't try to make up an answer.
        ------
        {context}
        """,
    ),
    ("human","{question}"),
])
chain = {"context": map_chain, "question": RunnablePassthrough()} | final_prompt | llm

# 6. chain.invoke("query")
chain.invoke("죽은 자가 산 자를 구할 수 있는가?")
'''
````

* def map_docs(inputs)에 대한 해설:
    documents = inputs['documents']
    question = inputs['question']
    results = []
    for document in documents:
        result = map_doc_chain.invoke({ # result = AIMessage
            "context": document.page_content,
            "question": question
        }).content
        results.append(result)
    results = "\n\n".join(results) # make it into one string
    return results
