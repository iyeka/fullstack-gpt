# Local embedding and LLM chat models.
1. HuggingFace
    1. HuggingFace Inference API
        - Using HuggingFace server.
        - Create api token and save it to .env file.
    2. Inference Endpoints: Using paid servers.
    3. Download model and run on computer
2. GPT4All
    1. 사이트에서 직접 모델 다운로드
3. Ollama
    - Nico's favourite.
    1. Download Ollama from website.
    2. Type ollama run model_name in Terminal.
    - How Ollama works.
    1. makes a request를 보낼 수 있는 server in your computer.
    2. Terminal에 [API EXAMPLE CODE](https://ollama.com/library/mistral:7b) 입력

```python
# 1-1. HuggingFace Inference API
from langchain.llms import HuggingFaceHub
from langchain.prompts import PromptTemplate

prompt = PromptTemplate.from_template("[INST]What is the meaning of {word}[/INST]") # follow the documentation of the model.
llm = HuggingFaceHub(
    repo_id="mistralai/Mistral-7B-Instruct-v0.1",
    model_kwargs = {
        "max_new_tokens":250, # Customize how many tokens we want to generate. 설정해야 긴 대답을 얻을 수 있다.
    }
)
chains = prompt | llm
chains.invoke("word":"potato")
```

```python
# 1-3. Download model and run on computer
from langchain.llms.huggingface_pipeline import HuggingFacePipeline
from langchain.prompts import PromptTemplate

llm = HuggingFacePipeline.from_model_id(
    model_id="gpt2",
    task="text-generation",
    device=0, # 0: gpu, -1: cpu
    pipeline_kwargs={"max_new_tokens": 150},
)

prompt = PromptTemplate.from_template("A {word} is a")

chain = prompt | llm
chain.invoke({"word":"potato"})
```

```python
# 2. GPT4All
from langchain.llms import GPT4All

llm = GPT4All(
    model="./#8_files/falcon.bin", # path to save
    device="gpu",
)

prompt = PromptTemplate.from_template("You are a helpful assistant that defines words. Define this word: {word}.")

chain = prompt | llm
chain.invoke({"word":"tomato"})
```
