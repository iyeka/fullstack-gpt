```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts import ChatPromptTemplate
from langchain.callbacks import StreamingStdOutCallbackHandler

chat = ChatOpenAI(
    temperature=0.4,
    streaming=True,
    callbacks=[StreamingStdOutCallbackHandler()],
)

poet_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are specialized in writing Haikus about programming languages."),
    ("human", "Write a Haiku about {programming_language}."),
])
critic_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are specialized in explaining Haikus."),
    ("human", "{haiku}"),
])

poet_chain = poet_prompt | chat
critic_chain = critic_prompt | chat
final_chain = {"haiku": poet_chain} | critic_chain

final_chain.invoke({"programming_language": "Ruby"})
```
