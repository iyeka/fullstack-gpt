## LCEL Chains:
- [x] Make a chain that is specialized in writing Haikus about programming languages and another one that is specialized in explaining Haikus.
    - [x] 1. 특정 프로그래밍 언어에 대한 시를 생성하도록 하는 [프롬프트 템플릿](https://python.langchain.com/v0.1/docs/modules/model_io/prompts/quick_start/#chatprompttemplate)을 만들고 해당하는 체인을 구축합니다.
    - [x] 2. 특정 시에 대한 설명을 생성하도록 하는 프롬프트 템플릿을 만들고 해당하는 체인을 구축합니다.
- [x] Chain the two chains together using [LCEL](https://python.langchain.com/v0.1/docs/expression_language/get_started/).
    - [x] 3. [두 개의 체인](https://python.langchain.com/v0.1/docs/expression_language/cookbook/multiple_chains/)을 LCEL을 사용하여 연결 합니다.
- [x] The final chain should receive the name of a programming language and it should reply with the Haiku and its explanation.
    - [x] 4. 답변이 생성되는 과정을 실시간으로 확인하기 위해 ChatOpenAI 모델에 streaming 옵션과 StreamingStdOutCallbackHandler을 활용하세요.
- [x] Use "gpt-3.5-turbo" as the model and ChatPromptTemplate for your prompts.

```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts import  ChatPromptTemplate
from langchain.callbacks import StreamingStdOutCallbackHandler

chat = ChatOpenAI(
    temperature=0.7,
    streaming=True,
    callbacks=[StreamingStdOutCallbackHandler()],
)

poet_prompt = ChatPromptTemplate.from_messages([
    ("system","You are a world-class poet who is specialized in writing Haikus about programming languages. Create a Haiku with the human given keyword."),
    ("human","{programming_language}"),
])
poet_chain = poet_prompt | chat

critique_prompt = ChatPromptTemplate.from_messages([
    ("system","You are a world-class critique who is specialized in explaining Haikus. Explain the human given haiku which is about programming language."),
    ("human","{haiku}"),
])
critique_chain = critique_prompt | chat

final_chain = {"haiku": poet_chain} | critique_chain
final_chain.invoke({"programming_language": "Ruby"})
```
