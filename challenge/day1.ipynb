    > - [x] 모델로는 "gpt-3.5-turbo"를 사용하고 프롬프트에는 ChatPromptTemplate을 사용하세요. 프롬프트 템플릿은 강의에서 다루었던 [ChatPromptTemplate](https://python.langchain.com/v0.1/docs/modules/model_io/prompts/quick_start/#chatprompttemplate) 를 활용하세요.
    > - [x] 답변이 생성되는 과정을 실시간으로 확인하기 위해 ChatOpenAI 모델에 streaming 옵션과 StreamingStdOutCallbackHandler을 활용하세요.

```python
from langchain.chat_models.openai import ChatOpenAI
from langchain.prompts import ChatPromptTemplate
from langchain.callbacks import StreamingStdOutCallbackHandler

llm_poet = ChatOpenAI(
    model="gpt-3.5-turbo",
    temperature=0.8,
    streaming=True,
    callbacks=[StreamingStdOutCallbackHandler()],
)

llm_critic = ChatOpenAI(
    model="gpt-3.5-turbo",
    temperature=0.4,
    streaming=True,
    callbacks=[StreamingStdOutCallbackHandler()],
)

prompt_poet = ChatPromptTemplate.from_messages([
    ("system", "You are a prominent poet. You are talented at writing Haikus about programming languages."),
    ("human", "Write a Haiku about {programming_language} programming language."),
])

prompt_critic = ChatPromptTemplate.from_messages([
    ("system", "You are a prominent critic. Read Haiku and explain about it."),
    ("human", "{haiku}"),
])

chain_poet = prompt_poet | llm_poet
chain_critic = prompt_critic | llm_critic

chain = {"haiku": chain_poet} | chain_critic

chain.invoke({
    "programming_language": "Ruby"
})
```
